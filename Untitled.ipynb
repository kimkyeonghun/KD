```python
import sys
import keras as k
import numpy as np
import tensorflow as tf
from tensorflow.keras import datasets

print(sys.version)
print(tf.__version__)
```

```python
from keras.preprocessing.image import ImageDataGenerator

(train_x, train_y), (test_x, test_y) = datasets.cifar10.load_data()
```

```python
datagen = ImageDataGenerator(rotation_range = 15,horizontal_flip = True, vertical_flip= True, width_shift_range= 0.1, height_shift_range= 0.1, zoom_range = 0.1)
datagen.fit(train_x)
```

```python
train_x=train_x.astype("float32")  
test_x=test_x.astype("float32")
mean=np.mean(train_x)
std=np.std(train_x)
test_x=(test_x-mean)/std
train_x=(train_x-mean)/std

# labels
num_classes=10
train_y = k.utils.to_categorical(train_y, num_classes)
test_y = k.utils.to_categorical(test_y, num_classes)
```

```python
import matplotlib.pyplot as plt
# plotting helper function
def plothist_acc(hist):
    plt.plot(hist.history['accuracy'])
    plt.plot(hist.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
```

```python
# plotting helper function
def plothist_loss(hist):
    plt.plot(hist.history['loss'])
    plt.plot(hist.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input,ZeroPadding2D,Conv2D,MaxPooling2D,Flatten,Dense,Dropout,Activation,BatchNormalization
from tensorflow.keras.optimizers import Adam
from keras.callbacks import EarlyStopping


with tf.device('/gpu:0'):
    Input_layer = Input(shape=(32,32,3))

    Zero1_1 = ZeroPadding2D((1, 1))(Input_layer)
    conv1_1 = Conv2D(64, (3, 3),activation='relu')(Zero1_1)
    ba1_1= BatchNormalization()(conv1_1)
    drop1 = Dropout(0.3)(ba1_1)
    Zero1_2 = ZeroPadding2D((1, 1))(drop1)
    conv1_2 = Conv2D(64, (3, 3),activation='relu')(Zero1_2)
    ba1_2= BatchNormalization()(conv1_2)
    stage1 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(ba1_2)

    Zero2_1 = ZeroPadding2D((1, 1))(stage1)
    conv2_1 = Conv2D(128, (3, 3),activation='relu')(Zero2_1)
    ba2_1= BatchNormalization()(conv2_1)
    drop2 = Dropout(0.4)(ba2_1)
    Zero2_2 = ZeroPadding2D((1, 1))(drop2)
    conv2_2 = Conv2D(128, (3, 3),activation='relu')(Zero2_2)
    ba2_2= BatchNormalization()(conv2_2)
    stage2 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(ba2_2)
    
    Zero3_1 = ZeroPadding2D((1, 1))(stage2)
    conv3_1 = Conv2D(256, (3, 3),activation='relu')(Zero3_1)
    ba3_1= BatchNormalization()(conv3_1)
    drop3_1 = Dropout(0.4)(ba3_1)
    Zero3_2 = ZeroPadding2D((1, 1))(drop3_1)
    conv3_2 = Conv2D(256, (3, 3),activation='relu')(Zero3_2)
    ba3_2= BatchNormalization()(conv3_2)
    drop3_2 = Dropout(0.4)(ba3_2)
    Zero3_3 = ZeroPadding2D((1, 1))(drop3_2)
    conv3_3 = Conv2D(256, (3, 3),activation='relu')(Zero3_3)
    ba3_3 = BatchNormalization()(conv3_3)
    stage3 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(ba3_3)
    
    Zero4_1 = ZeroPadding2D((1, 1))(stage3)
    conv4_1 = Conv2D(512, (3, 3),activation='relu')(Zero4_1)
    ba4_1= BatchNormalization()(conv4_1)
    drop4_1 = Dropout(0.4)(ba4_1)
    Zero4_2 = ZeroPadding2D((1, 1))(drop4_1)
    conv4_2 = Conv2D(512, (3, 3),activation='relu')(Zero4_2)
    ba4_2 = BatchNormalization()(conv4_2)
    drop4_2 = Dropout(0.4)(ba4_2)
    Zero4_3 = ZeroPadding2D((1, 1))(drop4_2)
    conv4_3 = Conv2D(512, (3, 3),activation='relu')(Zero4_3)
    ba4_3= BatchNormalization()(conv4_3)
    stage4 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(ba4_3)
    
    Zero5_1 = ZeroPadding2D((1, 1))(stage4)
    conv5_1 = Conv2D(512, (3, 3),activation='relu')(Zero5_1)
    ba5_1= BatchNormalization()(conv5_1)
    drop5_1 = Dropout(0.4)(ba5_1)
    Zero5_2 = ZeroPadding2D((1, 1))(drop5_1)
    conv5_2 = Conv2D(512, (3, 3),activation='relu')(Zero5_2)
    ba5_2 = BatchNormalization()(conv5_2)
    drop5_2 = Dropout(0.4)(ba5_2)
    Zero5_3 = ZeroPadding2D((1, 1))(drop5_2)
    conv5_3 = Conv2D(512, (3, 3),activation='relu')(Zero5_3)
    ba5_3= BatchNormalization()(conv5_3)
    stage5 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))(ba5_3)
    
    
    flat = Flatten()(stage5)
    dense1 = Dense(4096,activation='relu')(flat)
    dropd1 = Dropout(0.5)(dense1)
    dense2 = Dense(4096,activation='relu')(dropd1)
    dropd2 = Dropout(0.5)(dense2)
    output = Dense(10,activation='softmax')(dropd2)
    
    model = keras.Model(inputs=Input_layer,outputs=output)
    
    adm = Adam(lr=0.0001,decay=0, beta_1=0.9, beta_2=0.999, epsilon=1e-08)
    early_stopping = EarlyStopping(monitor='val_accuracy',patience=10,mode='auto',restore_best_weights=True)
    model.compile(loss='categorical_crossentropy', metrics=['accuracy'],optimizer=adm)
    history = model.fit(train_x,train_y,batch_size = 32, epochs = 100,callbacks=[early_stopping],validation_data=(test_x,test_y),shuffle=True,)
```

```python
plothist_acc(history)
```

```python
plothist_loss(history)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input,ZeroPadding2D,Conv2D,MaxPooling2D,Flatten,Dense,Dropout,Activation,BatchNormalization
from tensorflow.keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Lambda

Teacher_logits = keras.Model(model.input,model.layers[-1].output)
Temperature = 1
T_layer = Lambda(lambda x:x/Temperature)(Teacher_logits.output)
Softmax_layer = Activation('softmax')(T_layer)
Teacher_soften = keras.Model(model.input,Softmax_layer)

y_train_new = Teacher_soften.predict(train_x)

# Hard+soft
y_train_new = np.c_[train_y,y_train_new]
y_test_new = np.c_[test_y,test_y]
```

```python
from tensorflow.keras.metrics import categorical_accuracy
from tensorflow.keras.losses import kullback_leibler_divergence as KLD_Loss, categorical_crossentropy as logloss
def KD_loss(y_true,y_pred,lambd=0.5,T=10.0):
    y_true,y_true_KD = y_true[:,:10],y_true[:,10:]
    y_pred,y_pred_KD = y_pred[:,:10],y_pred[:,10:]
    # Classic cross-entropy (without temperature) hard target
    CE_loss = logloss(y_true,y_pred)
    # KL-Divergence loss for softened output (with temperature) soft target Loss_Teacher
    KL_loss = T**2*KLD_Loss(y_true_KD,y_pred_KD)

    return (1-lambd)*CE_loss + lambd*KL_loss

def accuracy(y_true,y_pred):
    return categorical_accuracy(y_true,y_pred)
```

```python
from tensorflow.keras.layers import concatenate

with tf.device('/gpu:0'):
    Input_layer = Input(shape=(32,32,3))
    
    zero1 = ZeroPadding2D((1, 1))(Input_layer)
    conv1 = Conv2D(32,(5,5),activation='relu')(zero1)
    bat1 = BatchNormalization()(conv1)
    drop1 = Dropout(0.3)(bat1)
    stage1 = MaxPooling2D((2,2),strides=(1,1))(drop1)
    
    zero2 = ZeroPadding2D((1, 1))(stage1)
    conv2 = Conv2D(64,(5,5),activation='relu')(zero2)
    bat2 = BatchNormalization()(conv2)
    drop2 = Dropout(0.4)(bat2)
    stage2 = MaxPooling2D((2,2),strides=(1,1))(drop2)
    
    zero3 = ZeroPadding2D((1, 1))(stage2)
    conv3 = Conv2D(128,(5,5),activation='relu')(zero3)
    bat3 = BatchNormalization()(conv3)
    drop3 = Dropout(0.4)(bat3)
    stage3 = MaxPooling2D((2,2),strides=(1,1))(drop3)
    
    flat = Flatten()(stage3)
    dense1 = Dense(500)(flat)
    dropd = Dropout(0.5)(dense1)
    output2 = Dense(10)(dropd)
    
    probs = Activation("softmax")(output2)
    StudentModel = keras.Model(Input_layer,probs)
    adam = Adam(learning_rate = 0.00001, beta_1=0.9, beta_2=0.999)
    early_stopping = EarlyStopping(monitor='val_accuracy',patience=10,mode='auto',restore_best_weights=True)
    StudentModel.compile(optimizer= adam,loss=lambda y_true,y_pred: KD_loss(y_true, y_pred,lambd=0,T=0),metrics=[accuracy])
    history = StudentModel.fit(train_x,train_y,epochs=200,validation_split=0.2,callbacks=[early_stopping],batch_size=32,shuffle=True)
```

```python
from tensorflow.keras.metrics import categorical_accuracy
from tensorflow.keras.losses import categorical_crossentropy as logloss
def LWF_loss(y_true,y_pred,lambd=0.1,T=2):
    soft_x,hard_x = y_true
    soft_y,hard_y = y_pred
    # Classic cross-entropy (without temperature) hard target
    old_loss = logloss(soft_x,soft_y)
    # KL-Divergence loss for softened output (with temperature) soft target Loss_Teacher
    new_loss = logloss(hard_x,hard_y)

    return lambd*old_loss + new_loss

def accuracy(y_true,y_pred):
    return categorical_accuracy(y_true,y_pred)
```
